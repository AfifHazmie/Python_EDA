## Data Preparation and Cleaning

In this section, we will be cleaning the data by removing null values and changing necessary data type in order to make analysis process much more easier.



Importing Pandas and Numpy libraries

```python
import numpy as np
import pandas as pd
```

Reading dataset into Data Frame

```python
df = pd.read_csv('https://raw.githubusercontent.com/muhdimranh/dataset/main/ElectionResults2018State.csv')
df
```

Exploring number of rows and columns

```python
numcolumns = len(df.axes[1])
numrows = len(df.axes[0])

print ("Number of columns:",numcolumns,"\nNumber of rows:",numrows)
```

Viewing dataset info to see datatypes for attributes

```python
df.info()
```

Deleting commas and percentage symbol to convert numeric values from object dtype to int and float dtype.

```python
df["Votes for Candidate"] = df["Votes for Candidate"].str.replace(',', '')
df["Total Votes Cast"] = df["Total Votes Cast"].str.replace(',', '')
df["% of total Votes"] = df["% of total Votes"].str.replace('%', '')
```

```python
df
```

Converting numeric columns from object dtype to int or float dtype.

```python
df["Votes for Candidate"] = pd.to_numeric(df["Votes for Candidate"], errors='coerce')
df["Total Votes Cast"] = pd.to_numeric(df["Total Votes Cast"], errors='coerce')
df["% of total Votes"] = pd.to_numeric(df["% of total Votes"], errors='coerce')
df.info()
```

```python
df
```

Checking if there's null value in any cell.

```python
df.isna().sum()
```

```python
checknull = df[df.isna().any(axis=1)]
checknull
```

Fill null value (written as MTB - Menang Tanpa Bertanding) with 0 values.

```python
df = df.fillna(0)
df
```

```python
df.iloc[580]
```

```python
checknullafter = df[df.isna().any(axis=1)]
checknullafter
```

Converting the Votes for Candidate column from float to integer.

```python
df["Votes for Candidate"] = df["Votes for Candidate"].astype(int)
df.info()
```

```python
df
```

# Finding the average voters in each of the available state

```python
state = df.groupby('State').apply(lambda x: x['% of total Votes'].mean()).reset_index(name= "PercentVote").sort_values(by='PercentVote', ascending=True)
fig, ax = plt.subplots(figsize=(25, 6))
sns.barplot(x=state['State'], y=state['PercentVote'])
ax.set(xlabel ="State", ylabel = "Percentage Vote", title ='Average Percentage Voters by State')
plt.show()
```

# Summmarize the Candidate Win By Candidate Party

```python
fig, ax = plt.subplots(figsize=(30, 10))

# Calculate the sum of the candidate win for each party.
party_totals = df.groupby('Candidate Party')['Candidate Win'].sum().sort_values()

# Create the histogram.
plt.bar(party_totals.index, party_totals.values)

# Add labels to the x and y axes.
plt.xlabel("Candidate Party")
plt.ylabel("Candidate Win")
plt.title('Candidate Win by Party in State Seat')
plt.xticks(rotation=90)

# Show the plot.
plt.show()
```

#### **Q1:** What is the total vote received by each party?

```python
# Select the relevant columns from the dataframe
df_plot = df[['Candidate Party', 'Votes for Candidate']]

# Group the data by party and calculate the total number of votes received by each candidate
df_plot = df_plot.groupby(['Candidate Party']).sum().reset_index().sort_values(by='Votes for Candidate')
df_plot
```

```python
f, ax = plt.subplots(figsize=(30,17)) # set the size that you'd like (width, height)

# Select the relevant columns from the dataframe
df_plot = df[['Candidate Party', 'Votes for Candidate']]

# Group the data by candidate party and calculate the total number of votes received by each group
df_plot = df_plot.groupby('Candidate Party').sum().reset_index().sort_values(by='Votes for Candidate')

# Set the y-axis to the candidate parties and the x-axis to the number of votes received
y = df_plot['Candidate Party']
x = df_plot['Votes for Candidate']

# Set the labels for the x-axis and y-axis
plt.ylabel('Candidate Party')
plt.xlabel('Number of Votes')
plt.title('Number of vote for each party')


# Create the horizontal bar chart
plt.barh(y, x)

# Loop through each bar and add the value on top of the bar
for i, v in enumerate(x):
    plt.text(v + 0.1, i, str(v) , va='center')

# Show the plot
plt.show()
```

**Description**


> A inverted bar graph showing number of vote receive for each party in the 2018 election. PKR party received the highest vote while party BEBAS - RUMAH receive the lowest number of vote



#### **Q2:** In the election, what is the percentage of candidates from each gender and what are the total votes received by the two genders?

```python
# Select the relevant columns from the dataframe
df_plot = df[['Candidate Name', 'Gender']]

# Group the data by gender and count the number of candidates in each group
df_plot = df_plot.groupby('Gender').count().reset_index()

# Set the labels for the pie chart to the genders and the sizes to the number of candidates
labels = df_plot['Gender']
sizes = df_plot['Candidate Name']

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.0f%%', pctdistance=1.1, labeldistance=0.5, radius = 2)
# Show the plot
plt.show()
```

**Description**

> Above pie chart shows the following: 

1.   L as in Lelaki(Male) takes majority of the pie chart 
2.   The percentage of L as Candidate in the election is higher with 89% than P with 11%






```python
f, ax = plt.subplots(figsize=(30,10)) # set the size that you'd like (width, height)
# Select the relevant columns from the dataframe
df_plot = df[['Gender', 'Votes for Candidate']]

# Group the data by gender and calculate the total number of votes received by each group
df_plot = df_plot.groupby('Gender').sum().reset_index()

# Set the x-axis to the genders and the y-axis to the number of votes received
x = df_plot['Gender']
y = df_plot['Votes for Candidate']

# Set the labels for the x-axis and y-axis
plt.xlabel('Gender')
plt.ylabel('Number of Votes (millions)')

# Divide the y-axis values by 1 million to convert them to millions
y = y / 1000000

# Create the bar chart
plt.bar(x, y)

# Loop through each bar and add the value on top of the bar
for i, v in enumerate(y):
    plt.text(i, v + 0.1, str(v) +' millions',  ha='center')

# Show the plot
plt.show()
```

**Description**

> Bar chart showing the number of voters vote for male and female candidate.

#### Q4: What are the 5 best and worst performers with their respective state in the election?

```python
# Group the data by state and candidate party, and calculate the percentage of total votes received by each group
df_plot = df.groupby(['State', 'Candidate Party'])['Votes for Candidate'].sum().reset_index()

# Sort the data by percentage of total votes in descending order
df_plot = df_plot.sort_values(by='Votes for Candidate', ascending=False)

# Select the top 5 candidate parties by percentage of total votes
df_plot = df_plot.head(5)

# Set the labels for the pie chart to the candidate party
labels = df_plot['Candidate Party'] + ' ' + df_plot['State']

# Set the sizes for the pie chart to the percentage of total votes received
sizes = df_plot['Votes for Candidate']

# Set the colors for the pie chart
colors = ['#ff9999','#66b3ff','#99ff99']

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', radius = 2)

# Show the plot
plt.show()
```

**Description**

> Above pie chart showing a distribution of the best 5 performers by candidate party in each state

> PKR Selangor received the highest number of votes among other party and states

> Candidate Party ,PKR hold 3 place in the top 5 best performer when receiving high number of vote

#### **Q5:** How many candidate for each state seat with divided according to candidate party.

**Calculating the overall total seat**

```python
df_unique = df.drop_duplicates()

# Count the total number of rows for each state.
seat_state = df_unique.groupby('State')['Seat Name'].count().sort_values()
total_seat = df_unique['Seat Name'].count()
# Print the result.
print(seat_state)
print('Total state seat',total_seat)
```

**Present the Data on a bar chart**

```python
#Setting the plot size
fig, ax = plt.subplots(figsize=(30, 10))

# Group the data by state and seat, and count the number of candidates for each group.
candidate_counts = df.groupby(['State', 'Candidate Party'])['Seat Name'].count()

# Get the indices for the x-axis tick marks.
x_indices = range(len(candidate_counts))

# Get the values for the bar heights.
heights = candidate_counts.values

# Get the labels for the x-axis tick marks.
labels = candidate_counts.index

# Create the bar chart.
plt.bar(x_indices, heights)

# Add labels to the x-axis.
plt.xticks(x_indices, labels, rotation=90)

# Add labels to each bar.
for x, y in zip(x_indices, heights):
    plt.text(x, y, y, ha='center', va='bottom')

# Add a title to the plot.
plt.title("Number of Candidates by State and Seat")

# Show the plot.
plt.show()
```

**Description**

> Based on above visualization, political party Barisan National(BN) have the most seat in Sabah with the second highest is Pakatan Rakyat(PKR) in Perak. From a total of **1646 seats**, BN participate in **60** of **252** seats in Sabah while PKR participate **59** of **171** seats in Perak.
